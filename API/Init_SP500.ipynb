```python
import bs4 as bs
import pickle
import requests
import datetime as dt
import pandas_datareader as web
import os
import time
```

```python
def save_sp500_tickers():
    resp = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})
    
    tickers = []
    for row in table.find_all('tr')[1:]:
        ticker = row.find_all('td')[0].text
#        print(ticker)
        if not ticker.find('.') == -1:
            s = list(ticker)
            s[ticker.find('.')] = '-'
            ticker = "".join(s)
            print(ticker)
        tickers.append(ticker.strip('\n'))
        
        
    #We now save the whole list to disk through the "pickle" library
    with open('sp500tickers.pickle', 'wb') as f:
        pickle.dump(tickers,f)
    
    return tickers
```

```python
def getStockPriceData(folder = 'dataSP500', reload_sp500 = False):
    #Retrive tickers and verify folder structure
    if reload_sp500:
        tickers = save_sp500_tickers()
    else:
        with open('sp500tickers.pickle', 'rb') as f:
            tickers = pickle.load(f)
    if not os.path.exists(folder):
        os.makedirs(folder)
    
    #Setup time period
    start = dt.datetime(1990,1,1)
    end = dt.datetime.now()
    
    #Retrieve data
    for ticker in tickers:
        if not os.path.exists('{fname}/{name}.csv'.format(fname=folder, name=ticker)): 
#            print('I AM HERE NOW:', ticker)
            df = web.DataReader(ticker, 'yahoo', start, end)
            if df.shape[0]<400:
                print(ticker,"not included, shape = ", str(df.shape))
                continue       
            df.to_csv('{fname}/{name}.csv'.format(fname=folder, name=ticker))
            print(ticker,"saved, shape = ", str(df.shape))
#            time.sleep(0.5)
        else:
            print('Already have {}'.format(ticker))
        
```
